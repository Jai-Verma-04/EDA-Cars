```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
cars = pd.read_csv('Cars.csv')
```

```python
cars.head()
```

```python
cars.shape
```

```python
cars.isnull().sum()
```

```python
cars['Cylinders'].fillna(cars['Cylinders'].mean(), inplace = True)
```

```python
cars.isnull().sum()
```

```python
cars['Make'].value_counts()
```

```python
cars[cars['Origin'].isin(['Asia', 'Europe'])]['Origin'].value_counts()
```

```python
cars.head()
```

```python
cars = cars[~(cars['Weight'] > 4000)]
```

```python
cars.shape
```

```python
cars.head()
```

```python
cars['MPG_City']=cars['MPG_City'].apply(lambda x:x+3)
```

```python
cars.head()
```

```python
plt.figure(figsize=(5,5))
plt.title("Distribution of CAR Makes")
plt.style.use('seaborn')
plt.bar(cars['Make'].value_counts().keys(), cars['Make'].value_counts())
plt.xlabel('Make -->'); plt.ylabel('Number -->')
plt.show()
```

```python
plt.figure(figsize=(5, 5))
plt.bar(cars['Origin'].value_counts().keys(), cars['Origin'].value_counts(), width=0.7)
plt.title('Number of CARS by Origin')
plt.xlabel("Origin -->"); plt.ylabel("Number -->")
plt.yticks([i for i in range(90, 150) if i%10==0])
plt.show()
```
